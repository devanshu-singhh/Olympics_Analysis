```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('athlete_events.csv')  
 
# data.head() display first 5 entry
# data.describe  about model
# data.info give info about data
print(data.head(), data.describe(), data.info())
```

```python
# regions and country noc data csv file
regions = pd.read_csv('datasets_31029_40943_noc_regions.csv')
print(regions.head())
 
# merging to data and regions frame
merged = pd.merge(data, regions, on='NOC', how='left')
print(merged.head())
```

```python
# creating goldmedal dataframes
goldMedals = merged[(merged.Medal == 'Gold')]
print(goldMedals.head())
```

```python
plt.figure(figsize=(20, 10))
plt.title('Distribution of Gold Medals')
sns.countplot(data=goldMedals, x='Age', palette="rainbow")  # Using a color palette for better visual effect
plt.show()
```

```python
goldMedals = merged[(merged.Medal == 'Gold')]
print('The no of athletes is',
      goldMedals['ID'][goldMedals['Age'] > 50].count(), '\n')
print(goldMedals[goldMedals['Age'] > 50])
```

```python
masterDisciplines = goldMedals['Sport'][goldMedals['Age'] > 50]

# Create the figure
plt.figure(figsize=(20, 10))
plt.tight_layout()

# Plot the count of disciplines for athletes over 50
sns.countplot(x=masterDisciplines, palette="rainbow")  # You can try other palettes like 'Set1', 'Pastel1', etc.
plt.title('Gold Medals for Athletes Over 50')

# Show the plot
plt.show()
```

```python
womenInOlympics = merged[(merged.Sex == 'F') &
                         (merged.Season == 'Summer')]
print(womenInOlympics.head(10))
 
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenInOlympics, palette="rainbow")
plt.title('Women medals per edition of the Games')
plt.show()
```

```python
print(goldMedals.region.value_counts().reset_index(name='Medal').head())
 
totalGoldMedals = goldMedals.region.value_counts().reset_index(name='Medal').head(5)
g = sns.catplot(x="region", y="Medal", data=totalGoldMedals,
                height=6, kind="bar", palette="rainbow")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Medals per Country')
plt.show()
```

```python
MenOverTime = merged[(merged.Sex == 'M') &
                     (merged.Season == 'Summer')]
wlMenOverTime = MenOverTime.loc[MenOverTime['Sport'] == 'Weightlifting']
 
plt.figure(figsize=(20, 10))
sns.pointplot(x='Year', y='Weight', data=wlMenOverTime, palette='Set2')
plt.title('Weight over year for Male Lifters')
plt.show()
```

```python
# Group data by Sport and Country, then count the number of medals
top_countries_by_sport = merged.groupby(['Sport', 'Team'])['Medal'].count().reset_index()
top_countries_by_sport = top_countries_by_sport.sort_values(by='Medal', ascending=False)

# Display top 10
print(top_countries_by_sport.head(10))

# Plot the data
plt.figure(figsize=(20, 10))
sns.barplot(x='Sport', y='Medal', hue='Team', data=top_countries_by_sport.head(10), palette='rainbow')
plt.title('Top 10 Countries by Medals in Different Sports')
plt.show()
```

```python
# Group data by Sport and Sex, then count the number of medals
gender_medals = merged.groupby(['Sport', 'Sex'])['Medal'].count().reset_index()

# Plot the gender distribution across different sports
plt.figure(figsize=(20, 10))
sns.barplot(x='Sport', y='Medal', hue='Sex', data=gender_medals, palette='Set1')
plt.title('Medals by Gender Across Different Sports')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group athletes by their name and count the number of unique appearances (years)
athlete_appearances = merged.groupby('Name')['Year'].nunique().reset_index()
athlete_appearances.columns = ['Name', 'Appearances']

# Merge this data with medal information
athlete_medals = merged.groupby('Name')['Medal'].count().reset_index()
athlete_data = pd.merge(athlete_appearances, athlete_medals, on='Name', how='left')

# Plot the data
plt.figure(figsize=(20, 10))
sns.scatterplot(x='Appearances', y='Medal', data=athlete_data, hue='Medal', palette='coolwarm', size='Medal')
plt.title('Athlete Longevity vs Medal Wins')
plt.xlabel('Number of Olympic Appearances')
plt.ylabel('Number of Medals')
plt.show()
```
